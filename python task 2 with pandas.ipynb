# Performing Simple Data Cleaning Tasks

 Check for Missing Values

```python
missing_values = data.isnull().sum() # Check for missing values
missing_values[missing_values > 0]   # Display columns with missing values
```

Fill Missing Values

```python
data_filled = data.fillna(0) # Fill missing values with 0
data_filled.head() # Display the first few rows of the filled data
```

Drop Rows with Missing Values

```python
data_dropped = data.dropna() # Drop rows with missing values
data_dropped.head()  # Display the first few rows of the data with dropped rows                           
```

Remove Duplicate Rows

```python
duplicates = data.duplicated().sum() # Check for duplicate rows
data_no_duplicates = data.drop_duplicates() # Remove duplicate rows
data_no_duplicates.head() # Display the first few rows of the data without duplicates
```

# Basic data manipulation operations

```python
filtered_data = data[data['ChipRate'] > 20] # Filter the data based on specific conditions
# Example: Filter rows where 'ChipRate' is greater than 20
# Display the first few rows of the filtered data
filtered_data.head()
```

```python
sorted_data = data.sort_values(by='ChipRate', ascending=False) # Sort the data based on one or more columns
# Example: Sort by 'ChipRate' in descending order
sorted_data.head() # Display the first few rows of the sorted data
```

```python
grouped_data = data.groupby('Observation').mean() # Group the data by a specific column and perform aggregations
# Example: Group by 'Observation' and calculate the mean of each group
grouped_data.head() # Display the first few rows of the grouped data
```
